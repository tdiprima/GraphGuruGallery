# Declarative Sine Wave

Uses HoloViews to declaratively create and display a sine wave curve with Bokeh backend.

```python
import holoviews as hv
import numpy as np

hv.extension("bokeh")
```

```python
xs = np.linspace(0, np.pi * 5, 150)
ys = np.sin(xs)
curve = hv.Curve((xs, ys)).opts(title="Sine Wave", color="blue", width=700)
curve
```
